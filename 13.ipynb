```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df
```

```python
# Inputs => X
# Output => y
X = df[['area']] # 2D array
y = df['price']
```

```python
model = LinearRegression() # Create the model
model.fit(X, y)
```

```python
# Predict prices
df["predicted_price"] = model.predict(X)
df
```

```python
print(f"Price = {model.coef_[0]:.2f} * Area + {model.intercept_:.2f}")
```

```python
print(865.00 * 587 + 175000.00)
```

```python
model.predict([[587]])
```

```python
# Plot
plt.figure(figsize=(8,5))
plt.plot(df['area'], df['price'], 'o-', label='Actual Price', color='blue')
plt.plot(df['area'], df['predicted_price'], 'r--', label='Linear Regression', linewidth=2)

# Labels and Legends
plt.title('Area vs Price with Linear Regression')
plt.xlabel('Area (sq. ft.)')
plt.ylabel('Price (₹)')
plt.legend()
plt.grid(True)
```
