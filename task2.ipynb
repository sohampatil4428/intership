```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("house_data.csv")
data
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
plt.grid()
```

```python
data.head(5)
```

```python
data["area"].fillna(data["area"].mode()[0],inplace=True)
data["halls"].fillna(data["halls"].mode()[0],inplace=True)
data["bedrooms"].fillna(data["bedrooms"].mode()[0],inplace=True)
data["age"].fillna(data["age"].mode()[0],inplace=True)
data["attached_washroom"].fillna(data["attached_washroom"].mode()[0],inplace=True)
data["floors"].fillna(data["floors"].mode()[0],inplace=True)
data["price"].fillna(data["price"].mode()[0],inplace=True)
```

```python
model=LinearRegression()
x=data[["area",	"halls",	"bedrooms",	"age",	"attached_washroom",	"floors"]]
y=data["price"]
model.fit(x,y)
```

```python
#                  area	halls	bedrooms	age	attached_washroom	floors
p1=model.predict([[8688,   2,      2,        18,        1,             3]])
p1
```
